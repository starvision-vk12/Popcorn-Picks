```python
#since our genre data is lsit of dictionary --> [{}] format so we have to convert
#it to list [] format, so we will create helper function using
#since the obj is string and we want string indices to be integer so we use
# ast module in which 'literal_eval' function does coverting work
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
#now we got genres in list format
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
#now for cast we want top three names
def convertcast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'].apply(convertcast)
```

```python
movies['cast'] = movies['cast'].apply(convertcast)
```

```python
movies.head(2)
```

```python
movies['crew'][0]
#now we only want the director name for each movie, 
# whiich means no other jobs are imp
```

```python
#now for cast we want top three names
def fetch_director(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'].apply(fetch_director)
#here we got the directors name
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#so lets do text vectorization
# Convert a collection of text documents to a matrix of token counts.
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
#so as we can see our tokens contains some prblm i.e
# 'activist',
#  'activities', 
#  'activity',
# similar words but showing different features so we need to make them same
# using steaming process
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)        #converting again list to string
```
